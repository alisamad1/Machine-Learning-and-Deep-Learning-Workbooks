```python
import numpy as np
```

```python
lst_marks = [45,32,78,89,34,67,89,90,56,78,99, 100,21]
minimum, Q1, median, Q3, maximum = np.quantile(lst_marks,[0,0.25, 0.50, 0.75, 1.0])
```

```python
minimum, Q1, median
```

```python
maximum
```

```python
Q3
```

```python
IQR = Q3 - Q1
IQR
```

```python
lower_fence = Q1 - 1.5*(IQR)
higher_fence = Q3 + 1.5*(IQR)
lower_fence
```

```python
higher_fence
```

```python
import seaborn as sns
sns.boxplot(lst_marks)
```
