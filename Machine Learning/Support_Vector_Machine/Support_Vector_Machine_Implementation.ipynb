```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
## Lets creaye synthetic data points
from sklearn.datasets import make_classification
```

```python
x,y = make_classification(n_samples = 1000, n_features = 2, n_classes = 2, n_clusters_per_class = 1, n_redundant= 0)
```

```python
x
```

```python
y
```

```python
pd.DataFrame(x)[0]
```

```python
sns.scatterplot(pd.DataFrame(x)[0],pd.DataFrame(x)[1], hue = y)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test,y_train, y_test = train_test_split(x,y,test_size=0.25, random_state=42)
```

```python
from sklearn.svm import SVC
```

```python
svc = SVC(kernel='linear')
```

```python
svc.fit(x_train,y_train)
```

```python
## prediction
y_pred = svc.predict(x_test)
```

```python
svc.coef_
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
print(classification_report(y_test,y_pred))
print(confusion_matrix(y_test,y_pred))
```

```python
rbf = SVC(kernel = 'rbf')
```

```python
rbf.fit(x_train, y_train)
```

```python
y_pred1 = rbf.predict(x_test)
```

```python
print(classification_report(y_test,y_pred1))
print(confusion_matrix(y_test,y_pred1))
```

```python
polynomial = SVC(kernel = 'poly')
polynomial.fit(x_train, y_train)
y_pred2 = polynomial.predict(x_test)
print(classification_report(y_test,y_pred2))
print(confusion_matrix(y_test,y_pred2))
```

```python
sigmoid = SVC(kernel = 'sigmoid')
sigmoid.fit(x_train, y_train)
y_pred3 = sigmoid.predict(x_test)
print(classification_report(y_test,y_pred3))
print(confusion_matrix(y_test,y_pred3))
```

```python
sigmoid.intercept_
```

## HyperParameter Tuning with SVC

```python
from sklearn.model_selection import GridSearchCV
param_grid = {'C' : [0.1,1,10,100,1000],
              'gamma': [1,0.1,0.01,0.001],
              'kernel' : ['rbf']}
```

```python
grid = GridSearchCV(SVC(),param_grid= param_grid,refit=True,cv = 5, verbose = 3)
```

```python
grid.fit(x_train,y_train)
```

```python
grid.best_params_
```

```python
y_pred4 = grid.predict(x_test)
print(classification_report(y_test, y_pred4))
print(confusion_matrix(y_test,y_pred4))
```

```python
a = df[['total_bill', 'tip', 'sex', 'smoker', 'day', 'time', 'size']]
b = df['total_bill']
```

```python
## train test split
from sklearn.model_selection import train_test_split
a_train,a_test,b_train,b_test = train_test_split(a,b,test_size = 0.25, random_state = 42)
```

```python
a_train.head()
```

```python
## Feature Encoding (Label Encoding and OneHotEncoding)
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le1 = LabelEncoder()
le2 = LabelEncoder()
le3 = LabelEncoder()
```

```python
a_train['sex'] = le1.fit_transform(a_train['sex'])
a_train['smoker'] = le2.fit_transform(a_train['smoker'])
a_train['time'] = le3.fit_transform(a_train['time'])
```

```python
a_train.head()
```

```python
a_test['sex'] = le1.transform(a_test['sex'])
a_test['smoker'] = le2.transform(a_test['smoker'])
a_test['time'] = le3.transform(a_test['time'])
```

```python
a_test.head()
```

```python
## onehotEncodinh
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
```

```python
ct = ColumnTransformer(transformers = [('onehot',OneHotEncoder(drop='first'),[3])],remainder = 'passthrough')
```

```python
ct.fit_transform(a_train)
```

```python
import sys
import numpy
numpy.set_printoptions(threshold=sys.maxsize)
a_train = ct.fit_transform(a_train)
```

```python
a_test = ct.transform(a_test)
```

```python
a_test
```

```python
## Support Vector Regresion 
from sklearn.svm import SVR
svr = SVR()
```

```python
svr.fit(a_train,b_train)
```

```python
b_pred = svr.predict(a_test)
```

```python
from sklearn.metrics import r2_score, mean_absolute_error
print(r2_score(b_test,b_pred))
```
