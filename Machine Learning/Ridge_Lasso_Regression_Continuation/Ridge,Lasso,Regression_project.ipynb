The dataset is converted into two sets based on regions from 122th index , we can make a new column based on region.

```python
df.loc[: 122, "Region"] = 0
df.loc[122 :, "Region"] = 1
df_new = df
```

```python
df_new[["Region"]] = df_new[["Region"]].astype(int)
```

## change the required columns as integer datatype

```python
df_new.columns
```

```python
df_new.columns = df_new.columns.str.strip()
df_new[["month", "day", "Year", "Temprature", "RH", "Ws"]] = df_new[["month", "day", "Year", "Temprature", "RH", "Ws"]].astype(int)
```

## changing the other columns to float data datatypw


```python
object = [features for features in df_new.columns if df_new[features].dtypes=="0"]
```

```python
for i in object:
    if i!="Classes":
        df_new[i] = df_new[i].astype(float)
```
