## Diabetes Prediction

```python
from sklearn.datasets import load_diabetes
```

```python
dataset = load_diabetes()
```

```python
print(dataset)
```

```python
dataset
```

```python
print(dataset['DESCR'])
```

```python
import pandas as pd
df_diabetes = pd.DataFrame(dataset.data, columns=[ 'age',
'sex',
'bmi',
'bp',
's1',
's2',
's3',
's4',
's5',
's6'
])
```

```python
df_diabetes.head()
```

```python
X = df_diabetes
Y = dataset['target']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train , Y_test = train_test_split(X,Y,test_size = 0.25, random_state = 10)
```

```python
X_train.head()
```

```python
## Checking the corrwlation
X_train.corr()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(15,10))
sns.heatmap(X_train.corr(),annot = True)
```

```python
from sklearn.tree import DecisionTreeRegressor
regressor = DecisionTreeRegressor()
regressor.fit(X_train, Y_train)
```

## Hyper Parameter Tuning

```python
params1 = {
    'criterion' : ['squared_error','friedman_mse', 'absolute_error'],
    'splitter' : ['best','random'],
    'max_depth' : [1,2,3,4,5,10,15,20,25],
    'max_features' : ['auto','sqrt', 'log2']
}
```

```python
params1
```

```python
from sklearn.model_selection import GridSearchCV
```

```python

regressor = DecisionTreeRegressor()
```

```python
grid = GridSearchCV(regressor, param_grid= params1, cv = 5, scoring = 'neg_mean_squared_error')
```

```python
import warnings
warnings.filterwarnings('ignore')
grid.fit(X_train,Y_train)
```

```python
grid.best_params_
```

```python
y_pred = grid.predict(X_test)
```

```python
y_pred
```

```python
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
```

```python
print(r2_score(Y_test,y_pred))
print(mean_squared_error(Y_test,y_pred))
print(mean_absolute_error(Y_test,y_pred))
```

```python
selectModel = DecisionTreeRegressor(criterion = 'friedman_mse', max_depth = 4, max_features = 'log2',splitter = 'best')
```

```python
selectModel.fit(X_train,Y_train)
```

```python
## Visualize the decision tree
import matplotlib.pyplot as plt
from sklearn import tree
plt.figure(figsize = (15,10))
tree.plot_tree(selectModel, filled = True)
```
