## Data Visualization with Matlplotlib
Matplotlib is a powerful plotting library for Python that enables the creation of static,animated and interactive visulaizations. It is widely used for data visualization in data Science and analytics. In this lesson, we will cover the basics of Matplotlib, including creating various types of plots and customizing them

```python
import matplotlib.pyplot as plt
```

```python
X = [1,2,3,4,5]
Y = [1,4,9,16,25]
## Create a line plot
plt.plot(X,Y)
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Basic Line Plot")
plt.show()
```

```python
## creating a customize line plot
X = [1,2,3,4,5]
Y = [1,4,9,16,25]
plt.plot(X,Y, color = "red",linestyle='--')
```

```python
## creating a customize line plot
X = [1,2,3,4,5]
Y = [1,4,9,16,25]
plt.plot(X,Y, color = "red",linestyle='--', marker = '*')
```

```python
## creating a customize line plot
X = [1,2,3,4,5]
Y = [1,4,9,16,25]
plt.plot(X,Y, color = "red",linestyle='--', marker = '*', linewidth = 2)
```

```python
## creating a customize line plot
X = [1,2,3,4,5]
Y = [1,4,9,16,25]
plt.plot(X,Y, color = "red",linestyle='--', marker = '*', linewidth = 2, markersize = 9)
```

```python
## creating a customize line plot
X = [1,2,3,4,5]
Y = [1,4,9,16,25]
plt.plot(X,Y, color = "red",linestyle='--', marker = '*', linewidth = 2, markersize = 9)
plt.grid(True)
```

```python
## Creating Multiple Plots
## sample data
X = [1,2,3,4,5]
y1 = [1,4,9,16,25]
y2 = [1,2,3,4,5]
plt.figure(figsize=(9,5))
plt.subplot(1,2,1)
plt.plot(X,y1, color = 'green')
plt.title("Plot 1")
plt.subplot(1,2,2)
plt.plot(y1,X,color = 'red')
plt.title("Plot 2")
```

```python
## Creating Multiple Plots
## sample data
X = [1,2,3,4,5]
y1 = [1,4,9,16,25]
y2 = [1,2,3,4,5]
plt.figure(figsize=(9,5))
plt.subplot(2,2,1)
plt.plot(X,y1, color = 'green')
plt.title("Plot 1")
plt.subplot(2,2,2)
plt.plot(y1,X,color = 'red')
plt.title("Plot 2")
plt.subplot(2,2,3)
plt.plot(X,y2,color = "blue")
plt.title("Plot 3")
plt.subplot(2,2,4)
plt.plot(y2,X, color = 'black')
plt.title("Plot 4")
```

```python
## Bar plots
categories = ['A', 'B', 'C', 'D', 'E']
values = [5,7,3,8,6]

## create a bar plot
plt.bar(categories,values, color = "blue")
plt.xlabel("Categories")
plt.ylabel("Values")
plt.title("Bar Plot")
plt.show()
```

## Histograms
Histograms are used to represent the distribution of a dataset. They divide the data into bins and count the number of data points in each bin.

```python
data  = [1,2,3,2.2,3,3,3,3,4,4,4,5,5,5,5,5]
## create a hsitogram
plt.hist(data, bins = 5,color = "orange", edgecolor = "black")
```

```python
x = [1,2,3,4,5]
y = [2,3,4,5,6]
plt.scatter(x,y,color = 'blue', marker = '*')
```

```python
## pie chart
labels = ['A', 'B', 'C', 'D']
sizes = [30,20,40,10]
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']
explode = (0.2,0,0,0)
## creating a pie chart
plt.pie(sizes, labels = labels,explode = explode, colors = colors, autopct = "%1.1f%%", shadow = True)
```

```python
## sales data visualization
import pandas as pd
sales_data_df=pd.read_csv('sales_data_sample.csv',encoding='latin1')
sales_data_df.head(5)
```

```python
## Plot total sales by products
total_quantityOrder_by_orderlinenumber = sales_data_df.groupby("QUANTITYORDERED")["ORDERLINENUMBER"].sum()
print(total_quantityOrder_by_orderlinenumber)
```

```python
total_quantityOrder_by_orderlinenumber.plot(kind='bar',color="purple")
```

```python
## plot sales trend over ordernumber
sales_trend = sales_data_df.groupby("ORDERNUMBER")["QTR_ID"].sum()
print(sales_trend)
```
