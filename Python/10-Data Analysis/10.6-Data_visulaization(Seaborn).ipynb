## Data Visualization with Seaborn
Seaborn is a Python visualization library based on Matplotlib that provides a high-level interface for drawing attractive and informative statistical graphics. Seaborn helps in creating complex visualizations with just few lines of code.In this lesson we will cover the basics of Seaborn, Including creating various types of plots and customizing them.

```python
import seaborn as sns
```

```python
## basic plotting with seaborn
tips = sns.load_dataset('tips')
print(tips)
```

```python
tips.head(5)
```

```python
tips.tail(5)
```

```python
## creating a scatter plot
import matplotlib.pyplot as plt

sns.scatterplot(x='total_bill', y='tip', data = tips)
plt.title("Scatter Plot of Total Bill Vs Tip")
```

```python
## line Plot
sns.lineplot(x = 'size', y = 'total_bill', data = tips)
plt.title("Line plot of Size vs Total bill")
plt.show()
```

```python
## Categorical Plots
## Bar Plots
sns.barplot(x='day',y='total_bill',data = tips)
plt.title("bar plot of day vs total bill")
plt.show()
```

```python
## Box Plot
sns.boxplot(x = 'day', y = 'total_bill', data = tips)
plt.title("Box Plot of Day vs Total_bill")
plt.show()
```

```python
## Violin Plot
sns.violinplot(x='day', y = 'total_bill', data = tips)
plt.title("violin Plot of Day vs Total_bill")
plt.show()
```

```python
## Creating Histograms using Seaborn
sns.histplot(tips['total_bill'], bins = 10, kde = True)
```

```python
## Kde Plot - Kernel Density Estimator Plot
sns.kdeplot(tips['total_bill'],fill=True)
```

```python
## Pair plot - to see the relationship between differnet variables
sns.pairplot(tips)
```

```python
## HeatMap
corr = tips[['total_bill', 'tip', 'size']].corr()
corr
```

```python
sns.heatmap(corr, annot = True, cmap = 'coolwarm')
```

```python
## Customization of the plot
import pandas as pd
sales_df = pd.read_csv('sales_data_sample.csv', encoding = 'latin1')
sales_df.head(5)
```

```python
plt.figure(figsize=(10,6))
sns.barplot(x ="QUANTITYORDERED", y = "ORDERLINENUMBER",data = sales_df, estimator = sum)
plt.title("Total QuantityOrderd")
plt.xlabel("QUANTITYORDERED")
plt.ylabel("ORDERLINENUMBER")
plt.show()
```
